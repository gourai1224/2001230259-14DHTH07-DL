### Câu 16: Viết chương trình Pandas chuyển đổi cột đầu tiên thành Series.

Trong Pandas, thuộc tính `.ix` đã bị loại bỏ (deprecated) từ phiên bản 0.20.0 và hoàn toàn không còn trong các phiên bản mới hơn. Thay vào đó, chúng ta nên sử dụng `.iloc` (dựa trên vị trí số nguyên) hoặc `.loc` (dựa trên nhãn) để truy cập dữ liệu.

Để chuyển đổi cột đầu tiên của DataFrame thành Series, chúng ta có thể sử dụng `.iloc[:, 0]` hoặc truy cập trực tiếp bằng tên cột nếu có.

```python
# Tạo một DataFrame mẫu
data = {'Col_A': [10, 20, 30, 40],
        'Col_B': [1, 2, 3, 4],
        'Col_C': ['X', 'Y', 'Z', 'W']}
df = pd.DataFrame(data)

print("DataFrame gốc:\n", df)

# Chuyển đổi cột đầu tiên ('Col_A') thành Series bằng .iloc
# .iloc[:, 0] chọn tất cả các hàng và cột thứ 0 (cột đầu tiên)
first_column_as_series_iloc = df.iloc[:, 0]

print("\nCột đầu tiên thành Series (sử dụng .iloc):\n", first_column_as_series_iloc)
print("Loại của kết quả: ", type(first_column_as_series_iloc))

# Hoặc chuyển đổi bằng cách truy cập trực tiếp tên cột (nếu biết tên cột)
first_column_as_series_name = df['Col_A']
print("\nCột đầu tiên thành Series (sử dụng tên cột):\n", first_column_as_series_name)
print("Loại của kết quả: ", type(first_column_as_series_name))
```

```python
# Câu 17: Viết chương trình Pandas sắp xếp theo cột của DataFrame 4 x 3 ngẫu nhiên.
df_cau17 = pd.DataFrame(np.random.randint(0, 100, size=(4, 3)), columns=['Cot_A', 'Cot_B', 'Cot_C'])
print("DataFrame gốc:\n", df_cau17)

# Sắp xếp DataFrame theo cột 'Cot_A' (mặc định là tăng dần)
df_sorted_A = df_cau17.sort_values(by='Cot_A')
print("\nDataFrame sắp xếp theo Cot_A (tăng dần):\n", df_sorted_A)

# Sắp xếp DataFrame theo cột 'Cot_B' giảm dần
df_sorted_B_desc = df_cau17.sort_values(by='Cot_B', ascending=False)
print("\nDataFrame sắp xếp theo Cot_B (giảm dần):\n", df_sorted_B_desc)
```

```python
# Câu 18: Viết chương trình Pandas thay đổi index của một Series (dùng pandas.reindex).
series_cau18 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print("Series gốc:\n", series_cau18)

# Định nghĩa index mới
new_index = ['e', 'b', 'a', 'f']

# Thay đổi index của Series bằng reindex
series_reindexed = series_cau18.reindex(new_index)
print("\nSeries sau khi reindex:\n", series_reindexed)

# Lưu ý: Các phần tử có index mới không tồn tại trong Series gốc sẽ có giá trị NaN.
```

```python
# Câu 19: Viết chương trình Pandas tìm những phần tử trong Series x mà không có trong Series y (dùng hàm isin của pandas).
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])

print("Series sr1:\n", sr1)
print("\nSeries sr2:\n", sr2)

# Tìm các phần tử trong sr1 mà không có trong sr2
elements_not_in_sr2 = sr1[~sr1.isin(sr2)]

print("\nItems of sr1 not present in sr2:\n", elements_not_in_sr2)
```

```python
# Câu 20: Tìm hiểu phép toán numpy.union1d, numpy.intersect1d (hội, giao)
# và viết chương trình minh họa 2 hàm này trên 2 Series x, y ngẫu nhiên có 20 phần tử.
series_x = pd.Series(np.random.randint(0, 50, size=20))
series_y = pd.Series(np.random.randint(25, 75, size=20))

print("Series X:\n", series_x)
print("\nSeries Y:\n", series_y)

# Chuyển đổi Series sang mảng NumPy để sử dụng các hàm của NumPy
arr_x = series_x.values
arr_y = series_y.values

# Thực hiện phép hợp (union) bằng numpy.union1d
union_result = np.union1d(arr_x, arr_y)
print("\nKết quả phép hợp (Union of X and Y):\n", union_result)

# Thực hiện phép giao (intersection) bằng numpy.intersect1d
intersect_result = np.intersect1d(arr_x, arr_y)
print("\nKết quả phép giao (Intersection of X and Y):\n", intersect_result)
```

```python
#Câu 21: Viết chương trình Pandas thêm Series y vào Series x theo chiều dọc (vertical),
# chiều ngang (horizontal) (Tìm hiểu hàm append và concat của pandas).
# Tạo Series x
series_x = pd.Series(np.arange(10))
print("Series x:\n", series_x)

# Tạo Series y
series_y = pd.Series(list('pqrstuvwxy'))
print("\nSeries y:\n", series_y)

# Thêm series y vào series x theo chiều dọc (vertical)
# Sử dụng pd.concat với axis=0 (mặc định) hoặc .append (đã bị deprecated trong các phiên bản mới của Pandas)
vertical_concat = pd.concat([series_x, series_y], axis=0)
print("\nKết quả nối theo chiều dọc (vertical):\n", vertical_concat)

# Thêm series y vào series x theo chiều ngang (horizontal)
# Sử dụng pd.concat với axis=1
horizontal_concat = pd.concat([series_x, series_y], axis=1)
print("\nKết quả nối theo chiều ngang (horizontal):\n", horizontal_concat)
```
