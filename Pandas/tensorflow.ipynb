### 10. Tạo ngẫu nhiên hai tensor 1-chiều x, y có 5 phần tử. Trả về x + y nếu x < y, và x – y nếu x > y (element-wise)

```python
x = tf.random.uniform(shape=[5], minval=0, maxval=10, dtype=tf.float32)
y = tf.random.uniform(shape=[5], minval=0, maxval=10, dtype=tf.float32)

result_tensor = tf.where(tf.less(x, y), x + y, x - y)

print("x:", x)
print("y:", y)
print("Result (x+y if x<y, else x-y):", result_tensor)
```

### 12. Viết hàm tạo 2 tensor ngẫu nhiên có kích thước x = (3, 2, 2), y = (3, 2, 2). Viết hàm tính cộng, trừ, nhân, chia sau đó chuyển các hàm này vào trong graph.

```python
# Create two random tensors
x = tf.random.uniform(shape=[3, 2, 2], minval=0, maxval=10, dtype=tf.float32)
y = tf.random.uniform(shape=[3, 2, 2], minval=0, maxval=10, dtype=tf.float32)

print("Tensor x:\n", x)
print("Tensor y:\n", y)

# Define arithmetic operations as TensorFlow functions
@tf.function
def add_tensors(t1, t2):
  return tf.add(t1, t2)

@tf.function
def subtract_tensors(t1, t2):
  return tf.subtract(t1, t2)

@tf.function
def multiply_tensors(t1, t2):
  return tf.multiply(t1, t2)

@tf.function
def divide_tensors(t1, t2):
  # Add a small epsilon to avoid division by zero
  return tf.divide(t1, t2 + tf.keras.backend.epsilon())

# Perform operations using the graph functions
sum_tensor = add_tensors(x, y)
diff_tensor = subtract_tensors(x, y)
prod_tensor = multiply_tensors(x, y)
div_tensor = divide_tensors(x, y)

print("\nSum (x + y):\n", sum_tensor)
print("\nDifference (x - y):\n", diff_tensor)
print("\nProduct (x * y):\n", prod_tensor)
print("\nDivision (x / y):\n", div_tensor)
```

### 13. Xây dựng model để tính tổng hai tensor x, y

```python
class SumModel(tf.Module):
  def __init__(self, name=None):
    super().__init__(name=name)

  @tf.function
  def __call__(self, x, y):
    return tf.add(x, y)

# Create an instance of the model
sum_model = SumModel()

# Generate new random tensors for demonstration
x_model_sum = tf.random.uniform(shape=[2, 2], minval=0, maxval=5, dtype=tf.float32)
y_model_sum = tf.random.uniform(shape=[2, 2], minval=0, maxval=5, dtype=tf.float32)

print("x for sum model:\n", x_model_sum)
print("y for sum model:\n", y_model_sum)
print("Sum from model:\n", sum_model(x_model_sum, y_model_sum))
```

### 14. Xây dựng model để tính hiệu và tích hai tensor x, y

```python
class DiffProdModel(tf.Module):
  def __init__(self, name=None):
    super().__init__(name=name)

  @tf.function
  def calculate(self, x, y):
    difference = tf.subtract(x, y)
    product = tf.multiply(x, y)
    return difference, product

# Create an instance of the model
diff_prod_model = DiffProdModel()

# Generate new random tensors for demonstration
x_model_diff_prod = tf.random.uniform(shape=[2, 2], minval=5, maxval=10, dtype=tf.float32)
y_model_diff_prod = tf.random.uniform(shape=[2, 2], minval=0, maxval=5, dtype=tf.float32)

print("x for diff/prod model:\n", x_model_diff_prod)
print("y for diff/prod model:\n", y_model_diff_prod)

difference_result, product_result = diff_prod_model.calculate(x_model_diff_prod, y_model_diff_prod)
print("\nDifference from model:\n", difference_result)
print("\nProduct from model:\n", product_result)
```

### 15. Xây dựng model để tính chu vi hình chữ nhật P = (a +b) * 2

```python
class RectanglePerimeterModel(tf.Module):
  def __init__(self, name=None):
    super().__init__(name=name)

  @tf.function
  def calculate_perimeter(self, a, b):
    # Ensure a and b are treated as TensorFlow tensors
    a = tf.cast(a, tf.float32)
    b = tf.cast(b, tf.float32)
    perimeter = tf.multiply(tf.add(a, b), 2.0)
    return perimeter

# Create an instance of the model
perimeter_model = RectanglePerimeterModel()

# Example usage
length = tf.constant(10.0)
width = tf.constant(5.0)

perimeter = perimeter_model.calculate_perimeter(length, width)
print(f"Length (a): {length.numpy()}, Width (b): {width.numpy()}")
print(f"Perimeter P = (a + b) * 2: {perimeter.numpy()}")

# Another example with different values
length_2 = 7.5
width_2 = 3.2
perimeter_2 = perimeter_model.calculate_perimeter(length_2, width_2)
print(f"\nLength (a): {length_2}, Width (b): {width_2}")
print(f"Perimeter P = (a + b) * 2: {perimeter_2.numpy()}")
```

16.Xây dựng model để tính diện tích toàn phần hình hộp chữ nhật

```python
def model_dien_tich_hinh_hop(a, b, c):
    # Chuyển đổi đầu vào thành tensor float32
    a = tf.cast(a, tf.float32)
    b = tf.cast(b, tf.float32)
    c = tf.cast(c, tf.float32)

    # S_xq = (a + b) * c * 2
    s_xq = tf.multiply(tf.add(a, b), tf.multiply(c, 2.0))

    # S_tp = (a * b) * 2 + S_xq
    s_tp = tf.add(tf.multiply(tf.multiply(a, b), 2.0), s_xq)

    return s_tp

# Test thử
print(f"Diện tích toàn phần: {model_dien_tich_hinh_hop(5, 3, 4).numpy()}")
```

17. Xây dựng model chứng minh hai cạnh bên hình thang song song theo điều kiện:Điều kiện: $AD // BC \iff (AD = BC \text{ và } AB = CD)$

```python
def model_chung_minh_song_song(AD_val, BC_val, AB_val, CD_val):
    # Kiểm tra điều kiện AD = BC và AB = CD
    cond1 = tf.equal(AD_val, BC_val)
    cond2 = tf.equal(AB_val, CD_val)

    # Kết quả logic: AD // BC
    ad_parallel_bc = tf.logical_and(cond1, cond2)
    return ad_parallel_bc

# Test: Nếu AD=BC và AB=CD thì trả về True (AD // BC)
print(f"AD song song BC: {model_chung_minh_song_song(10, 10, 5, 5).numpy()}")
```

18. Chứng minh hai cạnh đáy hình thang bằng nhauĐiều kiện: $AB = CD \iff (AD = BC \text{ và } AD // BC)$.

```python
def model_chung_minh_day_bang_nhau(AD_len, BC_len, is_AD_parallel_BC):
    # Kiểm tra AD = BC
    cond1 = tf.equal(AD_len, BC_len)

    # Kết quả logic: AB = CD
    ab_equal_cd = tf.logical_and(cond1, is_AD_parallel_BC)
    return ab_equal_cd

# Test
print(f"AB bằng CD: {model_chung_minh_day_bang_nhau(10, 10, True).numpy()}")
```

19. Xây dựng model chứng minh NA = NC theo mô tả sau với AM, MB, AN, NC là đầu vào: Dựa trên mô tả: Nếu $MN // BC$ và $M$ là trung điểm $AB$ ($AM = MB$), thì $N$ là trung điểm $AC$ ($NA = NC$).

```python
def model_chung_minh_trung_diem(AM, MB, is_MN_parallel_BC):
    """
    Input: AM, MB (độ dài), is_MN_parallel_BC (Boolean)
    Output: Boolean (Xác nhận NA = NC)
    """
    # Điều kiện 1: M là trung điểm AB
    is_M_midpoint = tf.equal(AM, MB)

    # Nếu M là trung điểm VÀ MN // BC thì N là trung điểm
    is_N_midpoint = tf.logical_and(is_M_midpoint, is_MN_parallel_BC)

    return is_N_midpoint

# Giả sử AM=5, MB=5 và MN song song BC
result = model_chung_minh_trung_diem(tf.constant(5), tf.constant(5), tf.constant(True))
print(f"Kết luận NA = NC: {result.numpy()}")
```
